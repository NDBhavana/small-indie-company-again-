```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
#MockupMetric1: Avg # of shots missed by players vs Levels
```

```python
num_levels = 5
num_players = 100

# Generate random data for missing shots at each level
missing_shots = np.random.randint(0, 7, (num_levels, num_players))

# Calculate the most common number of missed shots at each level (Level 0 to Level 4)
most_common_missed_shots = [np.bincount(missing_shots[level]).argmax() for level in range(num_levels)]

# Labels for the X-axis
level_labels = [f'Level{level}' for level in range(num_levels)]

# Create the bar graph
plt.figure(figsize=(10, 6))
plt.bar(level_labels, most_common_missed_shots, color='blue', alpha=0.7)

# Set labels and titles
plt.xlabel('Level')
plt.ylabel('Most Common Missed Shots')
plt.title('Most Common Number of Missed Shots at Each Level')

# Customize Y axis labels
plt.yticks(np.arange(0, 10))

# Display the plot
plt.show()
```

```python
# Generate random data
np.random.seed(0)  # For reproducibility
num_samples = 100  # Number of data points
levels = np.random.choice([0, 1, 2, 3, 4], num_samples)
time_spent = np.random.randint(3, 60, num_samples)  # Random time spent between 10 and 600 seconds

# Create a DataFrame
data = pd.DataFrame({'Level Name': levels, 'Time Spent': time_spent})

# Group the data by the "Level Name" column and calculate the average time spent for each level
average_time = data.groupby('Level Name')['Time Spent'].mean()

# Define a colorful palette
colors = plt.cm.viridis(np.linspace(0, 1, len(average_time)))

# Plot the data with custom colors
average_time.plot(kind='bar', color=colors)
plt.title('Average Time Spent on Each Level')
plt.xlabel('Level Name')
plt.ylabel('Average Time Spent (seconds)')
plt.show()
```
